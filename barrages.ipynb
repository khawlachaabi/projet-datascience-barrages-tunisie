```python
import pandas as pd
# Charger le dataset
df = pd.read_csv("barrages.csv")

# Afficher un aperçu des 5 premières lignes
df.head()
```

```python
# Afficher les informations générales (types, non-null, etc.)
df.info()
```

```python
# Vérifier les valeurs manquantes
df.isna().sum()
```

```python
#Renommer les colonnes
df = df.rename(columns={'Nom_Fr': 'nom_barrage', 'apports': 'apports_journaliers','lachers': 'lachers_journaliers', 'stock': 'stock_actuel',
'moy_mois': 'moyenne_mensuelle_apports', 'cumul_annee_prec': 'cumul_apports_annee_prec', 'moy_periode': 'moyenne_periode',
'Cap_tot_act': 'capacite_totale_actuelle', 'Cote': 'altitude_niveau_eau','Cap_tot_init': 'capacite_totale_initiale','fonctionnel': 'etat',
'Annee_prod': 'annee_mise_en_service',})
df.head(3)
```

```python
# Créer un taux de remplissage (%)
df['taux_remplissage'] = (df['stock_actuel'] / df['capacite_totale_actuelle']) * 100
#supprimer la colonne "Nom_Ar"
df = df.drop(['Nom_Ar'], axis=1)
# Convertir la colonne 'date' en format datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
 # Vérifier les noms des barrages
df['nom_barrage'].unique()
```

```python
# Stock total actuel (tous barrages confondus)
stock_total = df.groupby('Date')['stock_actuel'].sum()

# Taux de remplissage moyen (tous barrages confondus)
taux_remplissage_moyen = df.groupby('Date')['taux_remplissage'].mean()

# Apports totaux journaliers (somme sur tous les barrages chaque jour)
apports_journaliers = df.groupby('Date')['apports_journaliers'].sum()

# Lâchers totaux journaliers
lachers_journaliers = df.groupby('Date')['lachers_journaliers'].sum()

df['année'] = df['Date'].dt.year

# Apports cumulés par année
apports_par_année = df.groupby('année')['apports_journaliers'].sum()

# Lâchers cumulés par année
lachers_par_année = df.groupby('année')['lachers_journaliers'].sum()

# Stock moyen par année
stock_moyen_par_année = df.groupby('année')['stock_actuel'].mean()
```
